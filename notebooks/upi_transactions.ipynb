### Validate business hypothesis using statistical tests

```python
# 1. ANOVA Test:to evaluate whether fraud occurrence differed significantly across Device Types (android, feature_phone, ios, tablet)
# Null Hypothesis: Fraud occurrence is same across all device types. 
# If p-value is less than 0.05 we reject Null Hypothesis, meaning at least one device type has a statistically different fraud occurrence than the others.
```

```python
devices = df_upi_transaction_history['device_type'].unique()

fraud_groups = [df_upi_transaction_history[df_upi_transaction_history['device_type']==d]['fraud_flag']
                for d in devices]

f_stat, p_value = st.f_oneway(*fraud_groups)
print("F-stat:", f_stat)
print("P-value:", p_val)
if p_value<0.05:
    print("We REJECT the null hypothesis")
else:
    print("We FAILED TO REJECT the null hypothesis")
```

**Conclusion</br>1. P-value 0.128 is greater than 0.05 so fraud occurrence do not differ significantly across device types</br>
2. This suggests that device type does not have a meaningful impact on fraud occurrence, and fraud risk appears to be uniformly distributed across platforms.</br>**

```python
# Null Hypothesis: There is no relationship between total merchant transactions and fraudulent transactions.
# Alternative Hypothesis: There is a positive relationship between total merchant transactions and fraudulent transactions (i.e., when transactions increase, fraud also increases).
```

```python
merchant_summary = df_upi_transaction_history.groupby('merchant_id').agg(total_transactions=('transaction_id','count'),
                fraud_transactions=('fraud_flag','sum')).reset_index()

corr, p_value = st.pearsonr(merchant_summary['total_transactions'], merchant_summary['fraud_transactions'])

print("P-value:", p_value)
print("Correlation:", corr)

if p_value < 0.05:
    print("We REJECT the null hypothesis")
    if corr > 0:
        print("Positive Correlation")
    else:
        print("Negative Correlation")
else:
    print("We FAILED TO REJECT the null hypothesis")
```

**Conclusion</br>1. P-value 0.0002 is less than 0.05 so there is a significant relationship between merchant transactions and fraud </br>
2. Correlation 0.163 > 0 there is a positive relationship between total merchant transactions and fraudulent transactions </br>**
